```
# String Composition Problem.
def GetStringComposition(genome, k):
  composition = []
  for i in range(len(genome)-k +1):
    pat = genome[i:i+k]
    composition.append(pat)
  return composition

def CheckGetStringComposition():
  # k = 5
  # genome = 'CAATCCAAC'
  file = open(main_directory+"Week1/sc1.txt", "r")
  lines = file.readlines()
  k = int(lines[0].strip())
  genome = lines[1].strip()
  composition = GetStringComposition(genome, k)
  txt = '\n'.join(composition)
  f = open(main_directory+"Week1/sc1S.txt",'w')
  f.write(txt)
  f.close()
  # for c in composition:
  #   print(c)

CheckGetStringComposition()
```

```
def GetList2String(strList):
  s = ''
  for i in strList:
    s += i
  return s

# String Spelled by a Genome Path Problem
def GetPath2Genome(paths):
  genome_list = []
  genome_list.append(paths[0])
  for i in range(1, len(paths)):
    genome_list.append(paths[i][-1])
  return GetList2String(genome_list)

def CheckGetPath2Genome():
  # genome_paths = ['ACCGA',
  #                 'CCGAA',
  #                 'CGAAG',
  #                 'GAAGC',
  #                 'AAGCT']
  genome_paths = []
  file = open(main_directory+"Week1/gp.txt", "r")
  lines = file.readlines()
  for line in lines:
    genome_paths.append(line.strip())
  genome = GetPath2Genome(genome_paths)
  print(genome)

CheckGetPath2Genome()
```

```
def GetPrefix(kmer):
  return kmer[:-1]

def GetSuffix(kmer):
  return kmer[1:]

# Overlap Graph Problem.
def GetOverlappedGraph(kmers):
  graph = {k:[] for k in kmers}
  for i,kmer in enumerate(kmers):
    for j,kmer2 in enumerate(kmers):
      if i == j:
        continue
      if GetSuffix(kmer) == GetPrefix(kmer2):
        graph[kmer].append(kmer2)
  return graph

def CheckGetOverlappedGraph():
  file = open(main_directory+"Week1/ogp1.txt", "r")
  lines = file.readlines()
  kmers = [line.strip() for line in lines]
  graph = GetOverlappedGraph(kmers)
  for g in graph.keys():
    if len(graph[g]) != 0:
      print('%s -> '%(g), end='')
      for n in graph[g]:
        if n == graph[g][-1]:
          print(n)
          break
        print('%s,'%(n), end='')

# CheckGetOverlappedGraph()
```

```
# De Bruijn Graph from a String Problem
def GetDeBruijnGraph(genome, k):
  graph = {}
  for i in range(len(genome)-k+1):
    kmer = genome[i:i+k]
    kmer_pref = GetPrefix(kmer)
    kmer_suff = GetSuffix(kmer)
    if kmer_pref not in graph.keys():
      graph[kmer_pref] = []
    graph[kmer_pref].append(kmer_suff)
  return graph

def CheckGetDeBruijnGraph():
  k = 3
  genome = 'TAATGCCATGGGATGTT'
  # file = open(main_directory+"Week1/db.txt", "r")
  # lines = file.readlines()
  # k = int(lines[0].strip())
  # genome = lines[1].strip()
  graph = GetDeBruijnGraph(genome, k)
  for node in graph.keys():
    print('%s -> '%(node), end='')
    for i,n in enumerate(graph[node]):
      if i == len(graph[node]) - 1:
        print(n)
        break
      print('%s,'%(n), end='')
    
# CheckGetDeBruijnGraph()
```

```
# DeBruijn Graph from k-mers Problem:
def GetDeBruijnGraphKmer(kmers):
  graph = {}
  for kmer in kmers:
    kmer_pref = GetPrefix(kmer)
    kmer_suff = GetSuffix(kmer)
    if kmer_pref not in graph.keys():
      graph[kmer_pref] = []
    graph[kmer_pref].append(kmer_suff)
  return graph

def CheckGetDeBruijnGraphKmer():
  file = open(main_directory+"Week1/dbk1.txt", "r")
  lines = file.readlines()
  kmers = [line.strip() for line in lines]
  graph = GetDeBruijnGraphKmer(kmers)
  for node in graph.keys():
    print('%s -> '%(node), end='')
    for i,n in enumerate(graph[node]):
      if i == len(graph[node]) - 1:
        print(n)
        break
      print('%s,'%(n), end='')

CheckGetDeBruijnGraphKmer()
```
