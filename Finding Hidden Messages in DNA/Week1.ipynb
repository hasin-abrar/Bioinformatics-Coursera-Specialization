1. **PatternCount(genome, pattern):** Returns the count of how many times a pattern exists in a genome sequence.
2. **FrequentWords(text, k):** Returns the k-mers with maximum frequency. *Reverse compliment is not considered.*
3. **GetReverseCompliment(genome):** Returns the reverse compliment of a genome.
4. **GetPatternIndices(genome, pattern):** Returns the indices in the genome where the given pattern exists.
5. **Pattern2Number(pattern):** Returns the equivalent number representation of the given pattern when the patterns are sorted lexicographically.
6. **GetFrequencyArray(genome, k):** Returns the frequency of the different k-mers that exists in the given genome.
7. **Number2Pattern(number, k):** Returns the equivalent pattern from the given number. Reverse method of Pattern2Number(pattern).
8.**List2String(l):** Returns a list from a given string.
9. **FindingFrequentWordsBySorting(genome, k, flag = 0, t = 1) :** Returns the set of k-mers with maximum frequency in the genome sequences. If flag value is not equal to 0, then the set of k-mers is only returned when the maximum frequency is at least as large as t. **More efficient than GetFrequencyArray**
10. **ClumpFinding(genome, k , L, t):** Returns the set of k-mers within the genome, which are present at least t times within a window size of L in the genome.
11. **BetterClumpFinding(genome, k, L, t):** Same as ClumpFinding but better.

```
def PatternCount(genome, pattern):
  cnt = 0
  for i in range(len(genome)- len(pattern) + 1):
    if genome[i:i+len(pattern)] == pattern:
      cnt += 1
  return cnt

file = open(main_directory+"data.txt", "r")
lines = file.readlines()
print(lines[1])
print(PatternCount(lines[0].strip(),lines[1].strip()))
```

```
def FrequentWords(text, k):
  frequentPatterns = set()
  count = []
  for i in range(len(text)-k+1):
    pattern = text[i:i+k]
    count.append(PatternCount(text, pattern))
  maxValue = max(count)
  for i in range(len(text)-k + 1):
    if count[i] == maxValue:
      frequentPatterns.add(text[i:i+k])
  return frequentPatterns

file = open(main_directory+"data_freq.txt", "r")
lines = file.readlines()
print(FrequentWords(lines[0].strip(), int(lines[1].strip())))
# print(FrequentWords('ACGTTGCATGTCGCATGATGCATGAGAGCT', 4))
```

```
def GetReverseCompliment(genome):
  rc = []
  for x in genome[-1::-1]:
    if x == 'A':
      rc.append('T')
    elif x == 'T':
      rc.append('A')
    elif x == 'C':
      rc.append('G')
    else:
      rc.append('C')
  return rc


file = open(main_directory+"data_rc.txt", "r")
lines = file.readlines()

rc = ''.join(GetReverseCompliment(lines[0].strip()))
print(rc)
```

```
def GetPatternIndices(genome, pattern):
  k = len(pattern)
  indices = []
  for i in range(len(genome) - k + 1):
    if genome[i:i+k] == pattern:
      indices.append(i)
  return indices      

file = open(main_directory+"Vibrio_cholerae.txt", "r")
lines = file.readlines()

for i in GetPatternIndices(lines[0].strip(), 'CTTGATCAT'):
  print(i,end= ' ')
```

```
# ComputingFrequencies(Text, k)
from collections import defaultdict

def Pattern2Number(pattern):
  base4 = ''
  for i in pattern:
    if i == 'A':
      base4 += '0'
    elif i == 'C':
      base4 += '1'
    elif i == 'G':
      base4 += '2'
    else:
      base4 += '3'
  base4 = int(base4)
  # print(base4)
  indx = 0
  decimal = 0
  while base4 != 0:
    n = base4 % 10
    decimal += n*(4**indx)
    base4 //=10
    indx += 1
  return decimal    


def GetFrequencyArray(genome, k):
  default_value = lambda : 0
  freq_array = defaultdict(default_value)
  for i in range(len(genome)-k+1):
    indx = Pattern2Number(genome[i:i+k])
    freq_array[indx] += 1
  return freq_array

file = open(main_directory+"Week1/fa.txt", "r")
lines = file.readlines()

genome = lines[0].strip()
k = int(lines[1].strip())
freq_array = GetFrequencyArray(genome, k)
for i in range(4**k):
  print(freq_array[i], end=' ')
```

```
def Number2Pattern(number, k):
  indx = 0
  base4 = []
  pattern = []
  while number != 0:
    n = number % 4
    base4.append(n)
    number //= 4

  if len(base4) < k:
    for i in range(k - len(base4)):
      base4.append(0)
  
  base4 = base4[::-1]
  for i in base4:
    if i == 0:
      pattern.append('A')
    elif i == 1:
      pattern.append('C')
    elif i == 2:
      pattern.append('G')
    else:
      pattern.append('T')
  return pattern

pattern = Number2Pattern(5789,7)
for i in pattern:
  print(i,end='')

```

```
# FindingFrequentWordsBySorting
def List2String(l):
  s = ''
  for i in l:
    s += i
  return s

def FindingFrequentWordsBySorting(genome, k, flag = 0, t = 1):
  freq_patterns = set()
  indx = []
  _count = []
  for i in range(len(genome)-k +1):
    pattern = genome[i:i+k]
    number = Pattern2Number(pattern)
    indx.append(number)
    _count.append(1)
  
  indx.sort()
  for i in range(1,len(indx)):
    if indx[i] == indx[i-1]:
      _count[i] = _count[i-1] + 1
  _max = max(_count)
  if flag == 1:
    if _max < t:
      return freq_patterns
  # if max >= t, then normal code will suffice
  for i in range(len(_count)):
    if _count[i] == _max:
      freq_patterns.add(List2String(Number2Pattern(indx[i], k)))
  return freq_patterns
genome = 'AAGCAAAGGTGGG'
k = 2
freq_patterns = FindingFrequentWordsBySorting(genome, k)
for pat in freq_patterns:
  for s in pat:
    print(s,end='')
  print()
```

```
# ClumpFinding
from collections import defaultdict
def ClumpFinding(genome, k , L, t):
  freq_patterns = set()
  default_val = lambda : 0
  clump = defaultdict(default_val)
  for i in range(len(genome)-L+1):
    window = genome[i:i+L]
    freq_pattern = FindingFrequentWordsBySorting(window, k, flag=1, t=t)
    freq_patterns |= freq_pattern
  return freq_patterns    


```

```
# BetterClumpFinding
from collections import defaultdict
def BetterClumpFinding(genome, k, L, t):
  freq_patterns = set()
  default_val = lambda : 0
  clump = defaultdict(default_val)
  window = genome[0:L]
  freq_array = GetFrequencyArray(window,k)
  for i in range(4**k):
    if freq_array[i] >= t:
      pat = List2String(Number2Pattern(i, k) )
      freq_patterns.add(pat)

  for i in range(1, len(genome)-L+1):
    first_pat = genome[i-1 : i-1 + k]
    indx = Pattern2Number(first_pat)
    freq_array[indx] -= 1
    last_pat = genome[i+L-k: i+L]
    indx = Pattern2Number(last_pat)
    freq_array[indx] += 1
    if freq_array[indx] >= t:
      pat = List2String(Number2Pattern(indx, k) )
      freq_patterns.add(pat)
  return freq_patterns
```
